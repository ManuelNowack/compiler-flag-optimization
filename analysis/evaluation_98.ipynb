```python
import glob
import math
import pandas as pd
```

```python
tuners = {
    "Fourier": "fourier",
    "LowDegree": "fourier_low_degree",
    "MonoTuner": "mono",
    "RandomTuner": "random",
    "SRTuner": "srtuner"}
data_mean = []
data_margin_of_error = []
for tuner, path in tuners.items():
    runtimes = []
    for file in glob.glob(f"../evaluation/{path}/n_098_budget_0500_??.csv"):
        df = pd.read_csv(file, index_col=0).transpose()
        runtimes.append(df[tuner])
    assert runtimes
    df = pd.DataFrame(runtimes)
    data_mean.append(df.mean(0))
    data_margin_of_error.append(1.96 * df.std(0) / math.sqrt(len(df.index)))
    # diff_low = df.mean(0) - df.min(0)
    # diff_high = df.max(0) - df.mean(0)
    # data_margin_of_error.append(diff_low.where(diff_low > diff_high, diff_high))
mean = pd.DataFrame(data_mean, index=tuners.keys())
margin_of_error = pd.DataFrame(data_margin_of_error, index=tuners.keys())
```
