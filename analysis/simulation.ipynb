```python
import collections
import pandas as pd
```

```python
success_fraction = collections.defaultdict(list)
speedup_learn = collections.defaultdict(list)
speedup_opt = collections.defaultdict(list)
for search_space in range(20, 121, 10):
    for budget in range(100, 1001, 100):
        path = f"../simulation/noise_5e-3/n_{search_space:03d}_budget_{budget:04d}_00.csv"
        df = pd.read_csv(path, index_col=0).transpose()
        assert len(df.columns) == 4
        assert df.columns[0] == "Default"
        assert df["Default"].min() == df["Default"].max() == 1.0
        assert df.columns[1] == "Train"
        assert df.columns[3] == "Optimal"
        try:
            assert repetitions == len(df.index)
        except NameError:
            repetitions = len(df.index)
        tuner = df.columns[2]
        success_count = (df[tuner] < df["Train"]).sum()
        success_fraction[search_space].append(success_count / repetitions)
        speedup_learn[search_space].append((df["Train"] / df[tuner]).mean())
        speedup_opt[search_space].append((df[tuner] / df["Optimal"]).mean())
```

```python
# Probability that the learned flags are better than the best flags from the training data
df = pd.DataFrame(success_fraction, index=range(100, 1001, 100))
df.to_csv("simulation_success.csv")
df
```

```python
# Speedup of the learned flags over the best flags from the training data
df = pd.DataFrame(speedup_learn, index=range(100, 1001, 100))
df.to_csv("simulation_speedup_learn.csv")
df
```

```python
# Speedup of the optimal flags over the learned flags
df = pd.DataFrame(speedup_opt, index=range(100, 1001, 100))
df.to_csv("simulation_speedup_opt.csv")
df
```
