```python
import collections
import pandas as pd
```

```python
alphas = {
    "1e-1": "regularization_1e-1",
    "1e-2": "noise_5e-3",
    "1e-3": "regularization_1e-3",
    "1e-4": "regularization_1e-4",
    "1e-5": "regularization_1e-5"}
validate_scores = collections.defaultdict(list)
speedup = collections.defaultdict(list)
for alpha, directory in alphas.items():
    for search_space in range(20, 101, 10):
        file = f"../simulation/{directory}/n_{search_space:03d}_budget_0500_00_validate_score.csv"
        df = pd.read_csv(file, index_col=0).transpose()
        validate_scores[search_space].append(df["Fourier"].mean())

        file = f"../simulation/{directory}/n_{search_space:03d}_budget_0500_00.csv"
        df = pd.read_csv(file, index_col=0).transpose()
        speedup[search_space].append((df["Train"] / df["Fourier"]).mean())
```

```python
pd.DataFrame(validate_scores, index=alphas.keys())
```

```python
# Speedup of the learned flags over the best flags from the training data
pd.DataFrame(speedup, index=alphas.keys())
```
