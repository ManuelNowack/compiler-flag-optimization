```python
import glob
import pandas as pd
```

```python
data_default = []
data_train = []
data_fourier = []
for file in glob.glob(f"../evaluation/fourier/n_020_budget_0500_??.csv"):
    df = pd.read_csv(file, index_col=0).transpose()
    data_default.append(df["Default"])
    data_train.append(df["Train"])
    data_fourier.append(df["Fourier"])
data_all = [pd.DataFrame(data_default).mean(),
            pd.DataFrame(data_train).mean(),
            pd.DataFrame(data_fourier).mean()]
df = pd.DataFrame(data_all, index=["Default", "Train", "Fourier"])
df
```

```python
df_train = pd.DataFrame(data_train).reset_index()
df_fourier = pd.DataFrame(data_fourier).reset_index()
(df_fourier < df_train).sum() / len(df_train.index)
```
